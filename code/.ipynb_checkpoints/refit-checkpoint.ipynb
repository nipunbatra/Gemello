```python
from nilmtk import *
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
ds = DataSet("/data/refit/refit.h5")
```

```python
len(ds.buildings)
```

```python
out = {}
out_wash = {}
out_tv = {}
daywise_wash = {}
for num, building in ds.buildings.iteritems():
    out[num]={}
    mains = building.elec.mains()
    
    mains_df = mains.load(sample_period=60).next()
    diff_15min = mains_df[('power','active')].diff().dropna()
    abs_diff_15min = diff_15min.abs()
    tot = len(abs_diff_15min)
    lt_500 = 1.*(abs_diff_15min<500).sum()/tot
    bet_500_1000 = 1.*((abs_diff_15min>=500)&(abs_diff_15min<1000)).sum()/tot
    gt_1000 = 1.*(abs_diff_15min>=1000).sum()/tot
    
    #out[num]["power"] = mains_df.resample("1M",how="sum")[('power','active')]
    mains_df["day"] = mains_df.index.dayofweek
    mains_daywise = mains_df.groupby("day").mean().values.flatten()
    out[num] = np.hstack([mains_daywise, lt_500, bet_500_1000, gt_1000])
    
    try:
        wm = building.elec['washing machine']
        out_wash[num] = wm.load(sample_period=60).next()['2014-3-1':'2015-2-28']
        temp =out_wash[num].copy()
        temp["day"] = temp.index.dayofweek
        temp_daywise = temp.groupby("day").mean().values.flatten()
        daywise_wash[num] = temp_daywise
        out_wash[num] = out_wash[num].resample("1M",how="sum")[('power','active')]
    except:
        pass

out_df = pd.DataFrame(out).T
out_df.columns = ["fraction_1","fraction_2","fraction_3","fraction_4","fraction_5","fraction_6","fraction_7",
                  "lt_500","bet_500_1000","gt_1000"]
```

```python
import statsmodels.api as sm

def decompose(df, freq=96):
    res = sm.tsa.seasonal_decompose(df.fillna(method='ffill').values, freq=freq)
    return pd.DataFrame({"seasonal":res.seasonal, "trend":res.trend, "obs":res.observed}, index=df.index)
```
